```python
import re
import pandas as pd

#function to make dict whose keys are words and values are counts of words
def increment_dict(d, key, increment=1): 
    if key in d.keys():
        d[key] += increment
    else:
        d[key] = increment
    return d

def dict_pairs_to_df(d, sortby='value', indexby='number'):
    """ returns pandas df of key value pairs in dict without recursion (i.e. dict can only be single key-value pairs, not nested).
    Permitted values for sortby argument are 'value', 'value descending', 'key', 'key descending'.
    Permitted values for indexby argument are 'number' or 'key'."""
    rdf = pd.DataFrame()
    for key in d.keys():
        rdf = rdf.append(pd.DataFrame({'key':[key], 'value':[d[key]]}), ignore_index=True)
    assert sortby in ['value', 'value descending', 'key', 'key descending']
    if sortby.find('descending') != -1:
        sortasc = False
    else:
        sortasc = True
    assert indexby in ['number', 'key', 'key descending']
    if sortby.find('value') != -1:
        rdf.sort('value', ascending=sortasc, inplace=True)
    else:
        rdf.sort('key', ascending=sortasc, inplace=True)
    if indexby == 'number':
        rdf.reset_index(drop=True, inplace=True)
    else:
        rdf.set_index('key', drop=False, inplace=True)
    return rdf

# note: dict_pairs_to_df is a snippet; this script only uses one of its possible output permutations.
```

Note on the length_times_count column: it is what you would expect from the name. It allows you to see which words contribute the most total amount of characters to the list, weighted equally between the length of the word and the number of times it appears. I have found it a useful (albeit made-up) metric for this procedure.

```python
df_counts = dict_pairs_to_df(word_counts)
df_counts.columns = [['word', 'count']]
def calc_len_times_count(df):
    return pd.Series({'length_times_count':len(df['word']) * df['count']})
df_counts = df_counts.merge(df_counts.apply(calc_len_times_count, axis=1), left_index=True, right_index=True)
df_counts.sort('length_times_count', ascending=False, inplace=True)
all_words = list(df_counts.word)
print df_counts.head(40)
```

```python
columns_output = column_names[:]
before_count = 0
after_count = 0
max_before = 0
max_after = 0

edited_replacements = {}     # ensures replacement word does not already appear
for key in replacements.keys():
    if replacements[key] not in all_words:
        edited_replacements[key] = replacements[key]

for i in range(len(columns_output)):
    before_count += len(columns_output[i])
    max_before = max(max_before, len(columns_output[i]))
    name_words = columns_output[i].split(' ')
    for j in range(len(name_words)):
        for key in edited_replacements.keys():
            if name_words[j] == key:
                name_words[j] = edited_replacements[key]
            if name_words[j] == key + ')':  # only ignored punctuation is parenthesis, but you can add more.
                name_words[j] = edited_replacements[key] + ')'
            if name_words[j] == '(' + key:
                name_words[j] = '(' + edited_replacements[key]
    columns_output[i] = (' ').join(name_words)
    after_count += len(columns_output[i])
    max_after = max(max_after, len(columns_output[i]))
    
print 'Replacement has reduced %d total characters to %d.' % (before_count, after_count)
print 'Longest column was %d characters; now it is %d characters.' % (max_before, max_after)
print 'Here are the first ten columns:'
print columns_output[:10]
```
